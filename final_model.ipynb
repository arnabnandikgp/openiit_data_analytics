```python
df = pd.read_csv("pairs-label-training.csv")
df
```

```python
file=open('text_and_id_sorted.txt','rU')
content = file.readlines()
```

```python
y = df['label']
X = df.drop('label',axis=1)
```

```python
mother_corpus = []
for line in range(len(content)):
    mod_line=content[line]
    review = re.sub('[^a-zA-Z]', ' ', mod_line)
    review = review.lower()
    review = review.split()
    review = [lemmatizer.lemmatize(word) for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    mother_corpus.append(review)
    
```

```python
mother_corpus
```

```python
corpus = []
#content = file.readlines()
for row in range(len(X.index)):
    value1 = X.iat[row, 1]
    value2 = X.iat[row, 0]
    line1 = mother_corpus[value1-1]
    line2 = mother_corpus[value2-1]
    final_line=line1+' '+line2
    corpus.append(final_line)
```

```python
corpus
```

```python
cv = TfidfVectorizer()
# X_train_final = np.array(X_train_final)
X_transformed = cv.fit_transform(corpus).toarray()
X_transformed
```

```python
X_transformed.shape
```

```python
X_train_mega, X_test, y_train_mega, y_test = train_test_split(
X_transformed,y , random_state=104,test_size=0.2, shuffle=True)
```

```python
X_train, X_val, y_train, y_val = train_test_split(
X_train_mega,y_train_mega , random_state=104,test_size=0.25, shuffle=True)
```

```python
model = Sequential([
    Dense(32, activation='relu', input_shape=(4366,)),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid'),
])
```

```python
model.compile(optimizer='sgd',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
hist = model.fit(X_train, y_train,
          batch_size=32, epochs=10,
          validation_data=(X_val, y_val))
```

```python
model.evaluate(X_test, y_test)#[1]
```
